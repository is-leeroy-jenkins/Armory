### üèóÔ∏è Classes and Objects

- Python‚Äôs OOP model is based on **class-based inheritance** and **dynamic typing**.

````python
class Person:
    def __init__(self, name: str):
        self.name = name
    def greet(self):
        return f"Hello, I'm {self.name}"

p = Person("Ada")
print(p.greet())
````

- Inheritance

- All classes derive from `object`. 

- Method resolution order (MRO) follows C3 linearization (`help(C)` shows hierarchy).



````python
class Employee(Person):
    def __init__(self, name, title):
        super().__init__(name)
        self.title = title
````

### Object Construction, Initialization, and Destruction

| Method                             | Purpose                                             | Example Trigger                          |
| ---------------------------------- | --------------------------------------------------- | ---------------------------------------- |
| `__new__(cls, *args)`              | Allocates a new instance (object creation).         | `MyClass()` allocates before `__init__`. |
| `__init__(self, *args)`            | Initializes a newly created instance.               | Constructor logic.                       |
| `__del__(self)`                    | Finalizer; called when object is garbage-collected. | Object destruction.                      |
| `__class_getitem__(cls, key)`      | Supports `Class[key]` typing syntax.                | `MyClass[int]`.                          |
| `__init_subclass__(cls, **kwargs)` | Hook called when a class is subclassed.             | `class Child(Base)` triggers it.         |
| `__post_init__()`                  | Dataclasses hook after initialization.              | In `@dataclass`.                         |


### Object Representation

| Method                          | Purpose                                                | Example Trigger                 |
| ------------------------------- | ------------------------------------------------------ | ------------------------------- |
| `__repr__(self)`                | Official string representation; should be unambiguous. | `repr(obj)`                     |
| `__str__(self)`                 | User-friendly string representation.                   | `str(obj)` or `print(obj)`      |

| --------------------------------- | ---------------------------------------------- | --------------------------- |

## üß© Object Model and Attribute Mechanics

- Every Python object stores metadata in an internal `__dict__` mapping of attribute names to values.

- Attributes are resolved through the `descriptor protocol` and the `method resolution order (MRO)`.

````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

p = Point(3, 4)
print(p.__dict__)
print(Point.__mro__)
````

- `__mro__` lists the linearized inheritance chain used when resolving attribute lookups.

- This dynamic lookup is why you can monkey-patch class attributes at runtime.


````python
Point.origin = Point(0, 0)
print(Point.origin.x, Point.origin.y)
````

## üîí Encapsulation and Access Control



- Python does not enforce privacy syntactically but uses naming conventions:

- Encapsulation is semantic rather than absolute; Python trusts the developer‚Äôs discipline.


````python
class Account:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

a = Account("Dana", 100)
a.deposit(50)
print(a._Account__balance)   # access via mangled name
````

## üîÑ Concurrency 

- Python supports multiple concurrency models.

- However, true **parallel execution** of Python bytecode is constrained by the **Global Interpreter Lock (GIL)** ‚Äî a mutex preventing concurrent access to Python objects by multiple native threads.

- This ensures memory safety but limits CPU-bound parallelism.

- To achieve concurrency, Python employs **threading** for I/O-bound tasks and **multiprocessing** for CPU-bound workloads.

- Additionally, `asyncio` offers cooperative multitasking for highly concurrent I/O operations without threads.



### üßµ Threading (I/O Concurrency)

- Threads share memory space and are best for tasks such as network requests, file I/O, or waiting for external resources.

````python
import threading
import time

def worker(name):
    print(f"Thread {name} starting")
    time.sleep(1)
    print(f"Thread {name} done")

threads = [threading.Thread(target=worker, args=(i,)) for i in range(3)]

for t in threads: t.start()
for t in threads: t.join()

print("All threads completed")
````

- `start()` begins execution; `join()` blocks until the thread completes.

- The `threading` module uses OS-level threads but only one executes Python bytecode at a time due to the GIL.

- Use it for **concurrency**, not **parallelism**.

### üßÆ Multiprocessing (True Parallelism)

- Each process maintains its own interpreter instance, circumventing the GIL entirely.

- Useful for CPU-intensive tasks such as numeric computation or data processing.

- Each process has independent memory space ‚Äî use `multiprocessing.Queue` or `Pipe` for interprocess communication.


````python
from multiprocessing import Process, cpu_count

def compute(x): return x*x

if __name__ == "__main__":
    print("Cores:", cpu_count())
    procs = [Process(target=compute, args=(i,)) for i in range(4)]
    for p in procs: p.start()
    for p in procs: p.join()
    print("Parallel computation complete")
````

### ‚ö° AsyncIO (Cooperative I/O Concurrency)

- `asyncio` uses **event loops** and **coroutines** for asynchronous, non-blocking operations.

- Instead of threads, it multiplexes I/O-bound tasks efficiently.

````python
import asyncio

async def fetch(url):
    print("Fetching", url)
    await asyncio.sleep(1)
    return f"Data from {url}"

async def main():
    tasks = [fetch(u) for u in ["A", "B", "C"]]
    results = await asyncio.gather(*tasks)
    print(results)

asyncio.run(main())
````

- Each `await` yields control to the event loop, enabling cooperative scheduling.

- Use for thousands of concurrent socket or network operations with minimal overhead.

### üï∏Ô∏è Comparison of Concurrency Models

| Model             | Mechanism    | Best for     | True Parallelism | Shared Memory     |
| ----------------- | ------------ | ------------ | ---------------- | ----------------- |
| `threading`       | OS threads   | I/O tasks    | ‚ùå                | ‚úÖ                 |
| `multiprocessing` | OS processes | CPU tasks    | ‚úÖ                | ‚ùå                 |
| `asyncio`         | Coroutines   | Network, I/O | ‚ùå                | ‚úÖ (single thread) |

#### Closures and Functional Currying

Closures are functions that contain other nested functions with state from outer function.

In Python, a common way to use them is to keep track of the state. In the example below, the outer function, attack_counter keeps track of counts of attacks. The inner fuction attack_filter uses the "nonlocal" keyword in Python3, to modify the variable in the outer function.

This approach is called "functional currying". It allows for a specialized function to be created from general functions. As shown below, this style of function could be the basis of a simple video game or maybe for the statistics crew of a mma match.

````python
#nonlocal cannot modify this variable
#lower_body_counter=5
def attack_counter():
    """Counts number of attacks on part of body"""
    lower_body_counter = 0
    upper_body_counter = 0
    #print(lower_body_counter)
    def attack_filter(attack):
        nonlocal lower_body_counter
        nonlocal upper_body_counter
        attacks = {"kimura": "upper_body",
           "straight_ankle_lock":"lower_body",
           "arm_triangle":"upper_body",
            "keylock": "upper_body",
            "knee_bar": "lower_body"}
        if attack in attacks:
            if attacks[attack] == "upper_body":
                upper_body_counter +=1
            if attacks[attack] == "lower_body":
                lower_body_counter +=1
        print(f"Upper Body Attacks {upper_body_counter}, Lower Body Attacks {lower_body_counter}")
    return attack_filter
````

#### Lazy Evaluated Functions (Generators)

A very useful style of programming is "lazy evaluation". A generator is an example of that. Generators yield an items at a time.

The example below return an "infinite" random sequence of attacks. The lazy portion comes into play in that while there is an infinite amount of values, they are only returned when the function is called.

````python
def lazy_return_random_attacks():
    """Yield attacks each time"""
    import random
    attacks = {"kimura": "upper_body",
           "straight_ankle_lock":"lower_body",
           "arm_triangle":"upper_body",
            "keylock": "upper_body",
            "knee_bar": "lower_body"}
    while True:
        random_attack = random.choices(list(attacks.keys()))
        yield random_attack
````

#### Decorators:   Functions that wrap other functions

##### Randomized Sleep Decorator

Another useful technique in Python is to use the decorator syntax to wrap one function with another function. In the example below, a decorator is written that adds random sleep to each function call. When combined with the previous "infinite" attack generator, it generates random sleeps between each function call.

````python
def randomized_speed_attack_decorator(function):
    """Randomizes the speed of attacks"""

    import time
    import random

    def wrapper_func(*args, **kwargs):
        sleep_time = random.randint(0,3)
        print(f"Attacking after {sleep_time} seconds")
        time.sleep(sleep_time)
        return function(*args, **kwargs)
    return wrapper_func
````

````python
@randomized_speed_attack_decorator
def lazy_return_random_attacks():
    """Yield attacks each time"""
    import random
    attacks = {"kimura": "upper_body",
           "straight_ankle_lock":"lower_body",
           "arm_triangle":"upper_body",
            "keylock": "upper_body",
            "knee_bar": "lower_body"}
    while True:
        random_attack = random.choices(list(attacks.keys()))
        yield random_attack
````

````python
for _ in range(5):
    print(next(lazy_return_random_attacks()))
````

##### Timing Decorator

Using a decorator to time code is very common

````python
from functools import wraps
from time import time

def timing(f):
    @wraps(f)
    def wrap(*args, **kw):
        ts = time()
        result = f(*args, **kw)
        te = time()
        print(f"fun: {f.__name__}, args: [{args}, {kw}] took: {te-ts} sec")
        return result
    return wrap
````

Using decorator to time execution of a function

````python
@timing
def some_attacks():
  attack = lazy_return_random_attacks()
  for _ in range(5):
    print(next(attack))

some_attacks()
````

````python
def randomized_speed_attack_decorator(function):
    """Randomizes the speed of attacks"""
    
    import time
    import random
    
    def wrapper_func(*args, **kwargs):
        sleep_time = random.randint(0,3)
        print(f"Attacking after {sleep_time} seconds")
        time.sleep(sleep_time)
        return function(*args, **kwargs)
    return wrapper_func
````

````python
@randomized_speed_attack_decorator
def lazy_return_random_attacks():
    """Yield attacks each time"""
    import random
    attacks = {"kimura": "upper_body",
           "straight_ankle_lock":"lower_body", 
           "arm_triangle":"upper_body",
            "keylock": "upper_body",
            "knee_bar": "lower_body"}
    while True:
        random_attack = random.choices(list(attacks.keys()))
        yield random_attack
````

## Creational Patterns



### üè≠ Factory Method

#### Summary

- Define an interface for creating an object, but let subclasses choose the implementation.

#### Key Points

- Decouples creation from usage.

- Ensures consistent object families.

- Useful when deciding object types at runtime.

````python
from abc import ABC, abstractmethod

class Dialog(ABC):
    @abstractmethod
    def create_button(self):
        pass

    def render(self):
        btn = self.create_button()
        btn.on_click()
        btn.render()

class HTMLButton:
    def render(self):
        print("<button>HTML Button</button>")
    def on_click(self):
        print("HTML Button clicked")

class WebDialog(Dialog):
    def create_button(self):
        return HTMLButton()

WebDialog().render()
````

### üè¢ Abstract Factory

#### Summary

- Create related objects without specifying concrete classes.

#### Key Points

- Produces consistent families of objects.

- Hides concrete class construction.

- Ideal for platform-dependent UI, plugins, or theming.

````python
from abc import ABC, abstractmethod

class GUIFactory(ABC):
    @abstractmethod
    def button(self): pass
    @abstractmethod
    def checkbox(self): pass

class MacFactory(GUIFactory):
    def button(self): return "Mac Button"
    def checkbox(self): return "Mac Checkbox"

class WinFactory(GUIFactory):
    def button(self): return "Win Button"
    def checkbox(self): return "Win Checkbox"

def render(factory: GUIFactory):
    print(factory.button(), factory.checkbox())

render(MacFactory())
````

### üß± Builder

#### Summary:

- Construct complex objects step-by-step.

#### Key Points:

- Separates construction from representation.

- Supports incremental and optional configuration.

- Useful for objects with many parameters.

````python
class Car:
    def __init__(self):
        self.parts = []

class CarBuilder:
    def __init__(self):
        self.car = Car()

    def add_wheels(self):
        self.car.parts.append("4 wheels")
        return self

    def add_engine(self):
        self.car.parts.append("V8 engine")
        return self

    def build(self):
        return self.car

car = CarBuilder().add_wheels().add_engine().build()
print(car.parts)
````

### üß¨ Prototype

#### Summary:

- Clone existing objects instead of building new ones.

#### Key Points:

- Avoids expensive re-initialization.

- Relies on shallow/deep copying.

````python
import copy

class Doc:
    def __init__(self, text):
        self.text = text

d1 = Doc("Original")
d2 = copy.deepcopy(d1)
d2.text = "Copy"

print(d1.text, d2.text)
````

### üîÅ Singleton

#### Summary:

- Ensure a class has only one instance.

#### Key Points:

- Shared global state.

- Useful for caches, registries, logging.

- Python uses __new__ for enforcement.

````python
class Singleton:
    _instance = None
    def __new__(cls):
        if not cls._instance:
            cls._instance = super().__new__(cls)
        return cls._instance

print(Singleton() is Singleton())  # True
````

### üåâ Bridge

#### Summary:

- Separate abstraction from implementation.

#### Key Points:

- Change abstraction and implementation independently.

- Reduces subclass explosion.

- Useful for rendering, file formats, APIs.


````python
from abc import ABC, abstractmethod

class Renderer(ABC):
    @abstractmethod
    def circle(self, r): pass

class Vector(Renderer):
    def circle(self, r): print(f"Vector circle {r}")

class Raster(Renderer):
    def circle(self, r): print(f"Raster circle {r}")

class Circle:
    def __init__(self, renderer, radius):
        self.renderer = renderer
        self.radius = radius
    def draw(self):
        self.renderer.circle(self.radius)

Circle(Vector(), 5).draw()
````

### üå≥ Composite

#### Summary:

- Treat individual objects and compositions uniformly.

#### Key Points:

- Hierarchical structures (trees).

- Clients treat nodes and leaves the same.

- Common in UI, file systems, organization charts.

````python
class Node:
    def show(self): pass

class File(Node):
    def __init__(self, name): self.name = name
    def show(self): print(self.name)

class Folder(Node):
    def __init__(self, name):
        self.name = name
        self.children = []

    def add(self, node):
        self.children.append(node)

    def show(self):
        print(self.name)
        for child in self.children:
            child.show()

root = Folder("root")
root.add(File("notes.txt"))
root.show()
````

### ‚öôÔ∏è Flyweight

#### Summary:

- Share heavy objects efficiently.

#### Key Points:

- Reduces memory footprint.

## üé≠ Behavioral Pattern

### üîó Chain of Responsibility

#### Summary:

- Pass requests along a chain of handlers.

#### Key Points:

- Decouples sender from receiver.

- Stops at first handler able to process.

- Great for validation pipelines.

````python
class Handler:
    def __init__(self, next=None):
        self.next = next
    def handle(self, req):
        if self.next:
            self.next.handle(req)

class Auth(Handler):
    def handle(self, req):
        if req == "auth":
            print("Authenticated")
        else:
            super().handle(req)

Auth().handle("auth")
````

### üìù Memento

#### Summary:

- Save and restore object state.


#### Key Points:

- Undo/redo support.

- Snapshot-based recovery.

- Separates state from logic.

````python
class Editor:
    def __init__(self): self.text = \"\"
    def save(self): return self.text
    def restore(self, m): self.text = m

ed = Editor()
ed.text = \"draft\"
snap = ed.save()
ed.text = \"changed\"
ed.restore(snap)
print(ed.text)
````

### üëÅ Observer

#### Summary:

- Notify subscribers when a subject changes.
  
#### Key Points:

- Publish/subscribe pattern.

- GUI events and data streams use this model.

- Decouples producers from consumers.

````python
class Subject:
    def __init__(self):
        self.observers = []
    def add(self, o):
        self.observers.append(o)
    def notify(self, msg):
        for o in self.observers:
            o.update(msg)

class Logger:
    def update(self, msg):
        print("Log:", msg)

s = Subject()
s.add(Logger())
s.notify("Event happened")
````

### üé≠ State

#### Summary:

- Alter behavior based on internal state.
  
#### Key Points:

- Object appears to change class.

- Replaces long if/elif blocks.

- Useful for UI modes or workflow stages.

````python
from abc import ABC, abstractmethod

class State(ABC):
    @abstractmethod
    def handle(self): pass

class Happy(State):
    def handle(self): print("üôÇ")

class Sad(State):
    def handle(self): print("‚òπÔ∏è")

class Person:
    def __init__(self):
        self.state = Happy()
    def set(self, state):
        self.state = state
    def act(self):
        self.state.handle()

p = Person()
p.act()
p.set(Sad())
p.act()
````

### üéØ Strategy

#### Summary:

- Select algorithms dynamically at runtime.
  
#### Key Points:

- Algorithms are interchangeable.

- Removes conditional logic.

- Useful for pricing rules, routing, sorting.

````python
def add(a,b): return a+b
def sub(a,b): return a-b

def operate(strategy, a, b):
    return strategy(a,b)

print(operate(add, 5, 3))
````

### üß™ Template Method

#### Summary:

- Define algorithm skeleton; subclasses fill steps.

#### Key Points:

- Ensures consistent workflow.

- Subclasses customize specific phases.

- Common in frameworks.

````python
from abc import ABC, abstractmethod

class Meal(ABC):
    def cook(self):
        self.prepare()
        self.main()
        self.clean()
    @abstractmethod
    def prepare(self): pass
    @abstractmethod
    def main(self): pass
    def clean(self):
        print("Cleaning")

class Pasta(Meal):
    def prepare(self): print("Boiling water")
    def main(self): print("Cooking pasta")

Pasta().cook()
````

### üßç‚Äç‚ôÇÔ∏èüßç‚Äç‚ôÄÔ∏è Visitor

#### Summary:

- Add operations without modifying existing classes.
  
#### Key Points:

- Useful for traversing structures.

- Double dispatch pattern.

- Operations separated from data classes.

````python
class Dog:
    def accept(self, v): v.visit_dog(self)

class Cat:
    def accept(self, v): v.visit_cat(self)

class Visitor:
    def visit_dog(self, d): print("Visited Dog")
    def visit_cat(self, c): print("Visited Cat")

v = Visitor()
Dog().accept(v)
Cat().accept(v)
````

# Cascading Style Sheets (CSS)

HTML provides the structure and content of a web page (e.g., headings, paragraphs, images), while CSS dictates how these elements should be displayed visually. CSS rules are applied to HTML elements to modify their appearance, allowing for the creation of diverse and dynamic web designs.
- Styling: CSS controls the visual appearance of HTML elements, including properties like:
- Colors: Text, backgrounds, borders.
- Typography: Fonts, sizes, weights, line heights.
- Layout: Positioning of elements, spacing, columns.
- Decorative features: Animations, transitions, shadows, background images.

## üß± Document & Root Styling

- CSS controls presentation, layout, and visual hierarchy of HTML.
- Global styles usually start with the html and body selectors.
- The :root pseudo-class defines global custom properties (variables).
- The universal selector * applies reset or normalization rules.

```python
:root
{
  --font-body: 'Segoe UI', sans-serif;
  --color-bg: #f5f7fa;
  --color-text: #222;
  --color-accent: #3a7afe;
}

*
{
  margin: 0;
  padding: 0;
  box-sizing: border-box;
}

html, body
{
  font-family: var(--font-body);
  color: var(--color-text);
  background-color: var(--color-bg);
  line-height: 1.6;
}
```

- Flexbox (display: `flex`) and Grid (display: `grid`) simplify alignment.

# ‚öñÔ∏è CSS3 At-Rules

- At-rules are special CSS instructions that begin with @ and control global behavior like fonts, animations, feature support, imports, and responsive layouts.
- They do not style elements directly ‚Äî instead, they set conditions or configure rules that CSS engines must follow.
- Common at-rules: @media, @supports, @font-face, @keyframes, @import, @page, @layer.
- Crucial for responsive design, progressive enhancement, embedding external resources, and defining animations.
- At-rules can appear at the top-level of a stylesheet or inside other rules (depending on the rule).

## üñ•Ô∏è @media ‚Äî Responsive Design



- Allows CSS to adapt to screen sizes, device orientation, pixel density, and printing.
- Typically used to create mobile, tablet, desktop, and print styles.
- Targets breakpoints using conditions like max-width, orientation, and prefers-color-scheme.

```python
/* Base desktop/tablet styling */
.container {
  padding: 2rem;
  font-size: 1.2rem;
}

/* Mobile breakpoint: under 600px */
@media (max-width: 600px) {
  .container {
    padding: 1rem;
    font-size: 1rem;
  }
}

/* Dark mode preference */
@media (prefers-color-scheme: dark) {
  body {
    background: #111;
    color: #eee;
  }
}
```

## üß™ @supports ‚Äî Feature Queries

- Enables conditional styling based on browser feature support.
- Ideal for progressive enhancement with modern CSS capabilities.
- Replaces older browser-detection hacks with clean capability-based checks.

```python
/* Only apply if the browser supports grid layouts */
@supports (display: grid) {
  .card-container {
    display: grid;
    grid-template-columns: repeat(auto-fit, minmax(220px, 1fr));
    gap: 1rem;
  }
}

/* Fallback for older browsers */
@supports not (display: grid) {
  .card-container {
    display: flex;
    flex-wrap: wrap;
  }
}
```

## ‚úíÔ∏è @font-face ‚Äî Custom Fonts


- Allows you to load external or local fonts to use in your styles.
- Supports multiple file formats (woff2, woff, ttf, otf, svg).
- Provides fallback fonts if the custom font fails to load.

```python
@font-face {
  font-family: "Inter";
  src: url("fonts/Inter-Regular.woff2") format("woff2"),
       url("fonts/Inter-Regular.woff") format("woff");
  font-weight: 400;
  font-style: normal;
}

/* Usage */
body {
  font-family: "Inter", system-ui, sans-serif;
}
```

## üéûÔ∏è @keyframes ‚Äî Animations

- Defines named animations used by the animation: property.
- Specifies visual changes at percentage intervals (0‚Äì100%).
- Can animate transforms, opacity, color, size, and more.

```python
/* Animation definition */
@keyframes fadeInUp {
  0% {
    opacity: 0;
    transform: translateY(20px);
  }
  100% {
    opacity: 1;
    transform: translateY(0);
  }
}

/* Usage */
.card {
  animation: fadeInUp 0.7s ease-out forwards;
}
```

## üìÑ @import ‚Äî Importing Stylesheets

- Brings external stylesheets into the current one.
- Must appear before all other rules (except @charset).
- Useful for modular CSS architectures ‚Äî though avoided in performance-critical use.

```python
@import url("base/reset.css");
@import url("theme/colors.css");

/* Local rules here */
body {
  margin: 0;
  padding: 0;
}
```

## üñ®Ô∏è @page ‚Äî Print Formatting

- Used for controlling printed page appearance.
- Adjusts page margins, sizes, and bleed areas in print media.
- Ideal for printable reports, invoices, and documents.

```python
@page
{
  margin: 1in;
}

@media print
{
  h1
  {
    page-break-before: always;
  }
}
```

## üß© @layer ‚Äî Cascade Layers (Modern CSS)

- Introduces ordered layers for styling to control specificity battles.
- Helps large stylesheets avoid unpredictable overrides.
- Works like explicit z-index for CSS rules.

```python
/* Define layers */
@layer base, components, overrides;

@layer base
{
  button
  {
		  font-size: 1rem;
  }
}

@layer components {
  button.primary { background: blue; color: white; }
}

@layer overrides {
  button { font-size: 1.2rem; }
}
```

## üéöÔ∏è @property ‚Äî Registering Custom Properties

- Allows you to define typed CSS variables with inheritance rules.
- Enables animatable custom properties (not possible with unregistered vars).
- A powerful modern feature for advanced UI animations.

```python
@property --spin {
  syntax: "<angle>";
  inherits: false;
  initial-value: 0deg;
}

.spinner {
  animation: rotate 4s infinite linear;
}

@keyframes rotate {
  to { --spin: 360deg; transform: rotate(var(--spin)); }
}
```

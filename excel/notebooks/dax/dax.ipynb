- ğŸ”¹ `FILTER` builds row-level filtered tables  

- ğŸ”¹ `ALL` removes filters for total calculations  

- ğŸ”¹ `ALLEXCEPT` preserves chosen columns  

- ğŸ”¹ `KEEPFILTERS` makes filters additive  

### Examples


```

    

        CALCULATE(
            SUM(Sales[SalesAmount]),
            FILTER(Sales, Sales[SalesAmount] > 1000)
        )

        -- ALLEXCEPT: Remove All But Specific Filter
        Sales by Country, All Years :=
        CALCULATE(
            [Total Sales],
            ALLEXCEPT( Sales, Sales[Country] )
        )

        -- KEEPFILTERS: Additive Filters
        High Value US Sales :=
        CALCULATE(
            [Total Sales],
            FILTER( Sales, Sales[SalesAmount] > 1000 ),
            KEEPFILTERS( Customers[Country] = "USA" )
        )


```

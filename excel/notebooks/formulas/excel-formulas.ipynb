## ðŸ§© Formula Fundamentals

- Formulas start with = and follow `PEMDAS`; use parentheses for clarity.

- Relative references (A1) shift; absolute ($A$1) remain fixed; mixed ($A1/A$1) lock one axis.

- Favor Named Ranges and Tables for readability.

- Wrap potential errors in `IFERROR` or `IFNA`.

- Limit volatile functions such as `OFFSET`, `INDIRECT`, `RAND`, `NOW` in large models.

## ðŸ”¤ Text



- `LEN`: Returns the number of characters in a text string.

- `LEFT`: Extracts a specified number of characters from the beginning of a text string.

- `RIGHT`: Extracts a specified number of characters from the end of a text string.

- `MID`: Extracts a specified number of characters from the middle of a text string, requiring a starting position.

- `TRIM`: Removes extra spaces from a text string, leaving only single spaces between words.

- `UPPER`: Converts all letters in a text string to uppercase.

- `LOWER`: Converts all letters in a text string to lowercase.

- `PROPER`: Capitalizes the first letter of each word in a text string and converts the rest to lowercase.

- `SUBSTITUTE`: Replaces existing text with new text in a string.

- `FIND` and `SEARCH`: Locate the starting position of one text string within another.

- `CLEAN`: Removes non-printable characters from text. 

- Clean with `TRIM` and `CLEAN`; normalize case with `UPPER`, `LOWER`, `PROPER`.

- Slice using `LEFT`, `RIGHT`, `MID`; find with `SEARCH` or `FIND`.

- Replace with `REPLACE` or `SUBSTITUTE`.

- Combine with `CONCAT` or `TEXTJOIN`.

- Use `TEXT` for formatting at presentation time.

- `TEXTSPLIT`, `TEXTBEFORE`, and `TEXTAFTER `simplify tokenization.

## ðŸ” Lookup & Reference

- `VLOOKUP`: Searches for a value in the leftmost column of a table and returns a value in the same row from a specified column. 

- `HLOOKUP`: Searches for a value in the top row of a table and returns a value in the same column from a specified row. 

- `XLOOKUP`: A more modern and flexible function that can look up values in any column or row and return corresponding values. 

- `INDEX`: Returns a value or the reference to a value from within a table or range based on a given row and column number. 

- `MATCH`: Searches for an item in a range and returns the item's relative position (number) in that range

- Prefer `XLOOKUP` for bi-directional searches with optional defaults.

- `XMATCH` supports wildcard and approximate matching.

- Combine `INDEX` and `MATCH` for 2D lookups.

- `CHOOSECOLS` and `CHOOSEROWS` project subsets.

- `OFFSET` and `INDIRECT` are volatile; use only for dynamic referencing.

- `GETPIVOTDATA` extracts PivotTable values safely.

## ðŸ’° Financial

- `PMT`, `FV`, and `PV` compute loan payments and value of money.

- `RATE` and `NPER` solve for interest rate or term.

- `NPV`/`XNPV` and `IRR`/`XIRR` assess investment returns.

- `SLN`, `DDB`, and `DB` calculate depreciation.

- `PRICE` and `YIELD` evaluate bonds; `DURATION` and `MDURATION` measure sensitivity.

- `EFFECT` and `NOMINAL` reconcile compounding rates.

- `FV` (Future Value): Calculates the future value of an investment based on a constant interest rate and periodic payments. 

- `PV` (Present Value): Calculates the present value of an investment or loan. 

- `PMT` (Payment): Calculates the periodic payment for a loan or annuity based on constant payments and a constant interest rate. 

- `RATE`: Calculates the interest rate per period of an annuity. 

- `NPER` (Number of Periods): Calculates the number of periods for a loan or investment

- `IRR` (Internal Rate of Return): Calculates the internal rate of return for a series of cash flows. 

- `XIRR` (Internal Rate of Return for Irregular Cash Flows): Calculates the internal rate of return for a schedule of cash flows that is not necessarily periodic. 

- `NPV` (Net Present Value): Calculates the net present value of an investment based on a discount rate and a series of future payments. 

- `XNPV` (Net Present Value for Irregular Cash Flow Schedules): Calculates the net present value for cash flows occurring at irregular intervals. 

- `IPMT` (Interest Payment): Returns the interest payment for a given period of an investment. 

- `CUMIPMT`: Returns the cumulative interest paid on a loan between two periods. 

- `SLN` (Straight-Line Depreciation): Calculates depreciation using the straight-line method. 

- `DB` (Declining Balance): Calculates depreciation for a specified period using the declining-balance method
